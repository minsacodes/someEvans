Let me create a simple Data Frame by columns, using the **list**s I created before. For that I create a dictionary where the **key** is the name of column, and the **value** is the list:

```python
# now in a dict:
ListInDict={'names':namesL, 'ages':agesL, 'country':countryL, 'education':educationL}
# here it is:
ListInDict
```

This dictionary is not a data frame, but with pandas the conversion will be easy:

```python
students=pandas.DataFrame(ListInDict)
```

A very important operation is **subsetting**. Let me first get the categorical data levels:

```python
#what are our categories:
    
school.sctyp.value_counts()
```

```python
[school[COLUMN].value_counts() for COLUMN in school.iloc[:,1:6]]
```

* Basic subsetting:

```python
# then: only students from private schools
school[school.sctyp=='Private']
```

```python
# then: only MALE students from private schools
school[(school.sctyp=='Private') & (school.sex=='Male')]
```

* Using function **query()**:

```python
query1='sctyp=="Private"'
school.query(query1)
```

```python
query2='sctyp=="Private" & sex=="Male"'
school.query(query2)
```

* Using function **where()**:

```python
filter1=school.sctyp=="Private"
school.where(filter1)
```

```python
filter2=school.sex=="Male"
school.where(filter1 & filter2)
```

The function **where()** produces missing values. We need extra work:

```python
school.where(filter1 & filter2).dropna()
```
