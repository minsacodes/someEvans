```python
link='westcoastracedata.csv'

import pandas as pd

race=pd.read_csv(link)
```

```python
race
```

```python
# soma values have decimal, let's round UP using 'ceiling' function from numpy:

import numpy as np

race.iloc[:,2:8]=race.iloc[:,2:8].apply(np.ceil)
```

```python
# let's replace total population
race.iloc[:,2:8].sum(axis=1) # axis=0 will sum by column
```

```python
# then

race['total_population']=race.iloc[:,2:8].sum(axis=1)
```

```python
# divide each column by total and multiply by 100 to get SHARE in %:

race.iloc[:,2:8].div(race.total_population, axis=0)*100 # axis=0 by column
```

```python
# new data frame with percent
racepct=race.iloc[:,2:8].div(race.total_population, axis=0)*100
```

```python
# change column names
racepct.columns+'_pct'
```

```python
#then
racepct.columns=racepct.columns+'_pct'
```

```python
racepct
```

```python
# you could round:
racepct.round(2)
```

```python
racepct=racepct.round(2)
```

```python
# time to append
pd.concat([race,racepct],axis=1)
```

```python
# making change
raceall=pd.concat([race,racepct],axis=1)
```

```python
raceall
```

```python
#save result:

raceall.to_csv('raceall.csv')
```
